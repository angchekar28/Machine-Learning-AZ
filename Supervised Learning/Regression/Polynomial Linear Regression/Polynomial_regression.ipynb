# Polynomial Regression

## Importing the libraries

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Importing the dataset

```
df=pd.read_csv('Position_Salaries.csv')
X=df.iloc[:,1:-1].values #ignore column one and last one
y=df.iloc[:,-1].values #take last column
```

## Training the Linear Regression model on the whole dataset

```
from sklearn.linear_model import LinearRegression
lin = LinearRegression()
lin.fit(X, y)
```

## Training the Polynomial Regression model on the whole dataset

```
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=4)
X_Poly=poly.fit_transform(X)
lin2 = LinearRegression()
lin2.fit(X_Poly, y)
```

## Visualising the Linear Regression results

```
plt.figure(figsize=(6,3))
plt.scatter(X,y,color='red')
plt.plot(X,lin.predict(X),color='blue')
plt.title('Linear Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
```

## Visualising the Polynomial Regression results

```
plt.figure(figsize=(6,3))
plt.scatter(X,y,color='red')
plt.plot(X,lin2.predict(X_Poly),color='blue')
plt.title('Polynomial Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
```

## Visualising the Polynomial Regression results (for higher resolution and smoother curve)

```
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.figure(figsize=(6,3))
plt.scatter(X,y,color='red')
plt.plot(X_grid,lin2.predict(poly.fit_transform(X_grid)),color='blue')
plt.title('Polynomial Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
```

## Predicting a new result with Linear Regression

```
lin.predict([[6.5]])
```

## Predicting a new result with Polynomial Regression

```
lin2.predict(poly.fit_transform([[6.5]]))
```
